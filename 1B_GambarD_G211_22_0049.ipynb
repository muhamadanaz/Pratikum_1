```python
import numpy as np
import matplotlib.pyplot as plt

plt.axis([0, 150, 100, 0])
r = 40
alphal = np.radians(0)
alpha2 = np.radians(360)
dalpha = np.radians(2)
XC = 75
yc = 50
plt.scatter(XC, yc, s=10, color='k')  # Change xc to XC

for alpha in np.arange(alphal, alpha2, dalpha):
    x = XC + r * np.cos(alpha)
    y = yc + r * np.sin(alpha)
    plt.scatter(x, y, s=5, color='k')

plt.show()
```
